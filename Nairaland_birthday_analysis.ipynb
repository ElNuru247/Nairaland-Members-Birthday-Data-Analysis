# Using more analysis and visualization techniques for the nairaland birthday data generated from ***NairaLand_Members_Birthday_Data***

```python
import pandas as pd 
import matplotlib.pyplot as plt
from datetime import datetime as dt
%matplotlib inline
```

```python
#Reading the scrapped birthday data with pandas
bday_data=pd.read_csv('2018-08-24.csv')
bday_data.head()
```

## Analysing the data and finding insight

1) Find year celebrant was born 

2) Find the most common age

3) Find the less common age

4) Find the working class (25-59)

5) Find the retired (60 and above)

6) Find those still in secondary school (18 and below)

7) Find those in higher institutions or post secondary school (19-24)

8) Find celebrants born in same year (i.e same age)

```python
#Lets create a new column to store birth year
bday_data['Birth_Year'] = dt.now().year - bday_data['Age']
bday_data.head()
```

```python
# most common ages
most_common=bday_data['Age'].value_counts().head()
most_common
```

```python
# less common ages
less_common=bday_data['Age'].value_counts().tail()
less_common
```

```python
#working class (25-59)
working_class=bday_data[(bday_data['Age']>=25) & (bday_data['Age']<60)]
working_class.sort_values(by='Age')
```

```python
#retired (60 and above)
retired=bday_data[bday_data['Age'] >= 60]
retired.sort_values(by='Age')
```

```python
#still in secondary school (18 and below)
in_secondary_school=bday_data[bday_data['Age']<= 18]
in_secondary_school.sort_values(by='Age')
```

```python
# in higher institutions or post secondary school (18-24)
in_higher_institution=bday_data[(bday_data['Age'] > 18) & (bday_data['Age'] < 25)]
in_higher_institution.sort_values(by='Age')
```
